Function that allows user to input a state and outputs relevant radar station for each given state.

```python


def get_radar_stations_by_state(state_territory_code):


  # Construct the API URL for the radar stations.
  radar_stations_url = f"https://api.weather.gov/stations?state={state_territory_code}"

  try:
    response = requests.get(radar_stations_url)
    response.raise_for_status()
    radar_stations_data = response.json()

    # Create a dictionary mapping station IDs to details.
    radar_station_dict = {}
    if 'features' in radar_stations_data:
      for station in radar_stations_data['features']:
        if 'properties' in station and 'stationIdentifier' in station['properties']:
          station_id = station['properties']['stationIdentifier']
          radar_station_dict[station_id] = station

    return radar_station_dict

  except requests.exceptions.RequestException as e:
    print(f"Error fetching radar stations for {state_territory_code}: {e}")
    return None

# Get user input for the state/territory code.
state_territory_code = input("Enter a State/Territory Code (e.g., NY, CA): ")

# Get the radar stations for the specified state/territory.
radar_stations = get_radar_stations_by_state(state_territory_code)

if radar_stations:
  print(f"Radar stations for {state_territory_code}:")
  for station_id, station_data in radar_stations.items():
    print(f"- {station_id}: {station_data}")
else:
  print(f"No radar stations found for {state_territory_code}.")
```
